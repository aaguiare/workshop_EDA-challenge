```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import cufflinks as cf
cf.go_offline()
```

```python
city = pd.read_csv("./data/diamonds_city.csv")
city
```

```python
clarity = pd.read_csv("./data/diamonds_clarity.csv")
clarity
```

```python
color = pd.read_csv("./data/diamonds_color.csv")
color
```

```python
dimensions = pd.read_csv("./data/diamonds_dimensions.csv")
dimensions
```

```python
properties = pd.read_csv("./data/diamonds_properties.csv")
properties
```

```python
Transactional = pd.read_csv("./data/diamonds_transactional.csv")
Transactional
```

```python
cut = pd.read_csv("./data/diamonds_cut.csv")
cut
```

```python
first_merge = pd.merge(Transactional, city, on = "city_id")
first_merge
```

```python
second_merge = pd.merge(first_merge, properties, on = "index_id")
second_merge
```

```python
third_merge = pd.merge(second_merge, cut, on = "cut_id")
third_merge
```

```python
forth_merge = pd.merge(third_merge, dimensions, on = "index_id")
forth_merge
```

```python
fifth_merge = pd.merge(forth_merge, clarity, on = "clarity_id")
fifth_merge
```

```python
final_diamonddf = pd.merge(fifth_merge, color, on = "color_id")

final_diamonddf
```

```python
fig = px.scatter(final_diamonddf, y='price', x='carat', color = "clarity")
fig.show()
```

```python


grouped_df = final_diamonddf.groupby('city')['index_id'].count().reset_index()

grouped_df.columns = ['city', 'number_of_diamonds']


fig = px.bar(grouped_df, x='city', y='number_of_diamonds', title='Number of Diamonds per City')


fig.show()
```

```python

grouped_df = final_diamonddf.groupby('city')['price'].median().reset_index()

grouped_df.columns = ['city', 'price_of_diamonds']


fig = px.bar(grouped_df, x='city', y='price_of_diamonds', title='Price of Diamonds per City')


fig.show()
```

```python


grouped_df = final_diamonddf.groupby('city').agg(
    total_price=pd.NamedAgg(column='price', aggfunc='median'),
    total_volume=pd.NamedAgg(column='index_id', aggfunc='count')
).reset_index()


fig = px.bar(grouped_df, x='city', 
             y=['total_price', 'total_volume'], 
             barmode='group',
             title='Total Price and Volume of Diamonds per City')


fig.show()
```

```python
grouped_df = final_diamonddf.groupby('cut').agg(
    price =pd.NamedAgg(column='price', aggfunc='median'),
    total_volume=pd.NamedAgg(column='index_id', aggfunc='count')
).reset_index()


fig = px.bar(grouped_df, x='cut', 
             y=['price', 'total_volume'], 
             barmode='group',
             title='Average price and Volume of Diamonds per cut quality')


fig.show()
```
